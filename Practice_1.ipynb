```python
def get_dummies(x):
    global cars
    temp_df = pd.get_dummies(eval('cars'+'.'+x),drop_first=True)
    cars=pd.concat([cars,temp_df],axis=1)
    cars.drop([x],axis=1,inplace=True)
    
    return cars
```

```python
dummies_list = ['fueltype', 'aspiration', 'carbody','drivewheel','enginelocation', 'enginetype','cylindernumber']
```

```python
for i in dummies_list:
    get_dummies(i)
```
