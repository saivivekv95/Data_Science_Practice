```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#Loading data

cData=pd.read_csv('/content/auto-mpg.csv')
```

```python
cData['origin']=cData['origin'].replace({1:'America',2:'Europe',3:'Asia'})
```

```python
#Dropping unecessary columns
cData.drop('car name',axis=1)
```

```python
hpIsDigit=pd.DataFrame(cData.horsepower.str.isdigit())
```

```python
cData[hpIsDigit['horsepower'] == False]
```

```python
cData = cData.replace('?',np.nan)
cData[hpIsDigit['horsepower'] == False]
```

```python
cData.isnull().sum()
```

```python
cData.median()
```

```python
cData['horsepower']=cData['horsepower'].fillna(cData['horsepower'].median())
```

```python
cData['horsepower']=cData['horsepower'].astype('float64')
```

```python
cData.dtypes
```

```python
#Dummy variable for Origin
cData=pd.get_dummies(cData,columns=['origin'])
```

```python
cData
```

```python
cData.drop('car name',axis=1)
```

```python
cData=cData.drop('car name',axis=1)
```

```python
cData
```

```python
#Splitting the data for the regression
y=cData['mpg']
x=cData.drop('mpg',axis=1)
```

```python
x
```

```python
y
```

```python
#splitting the dataset for training and testing

X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
#Initialising the regression
regression_model=LinearRegression()
```

```python
regression_model.fit(X_train,Y_train)
```

```python
regression_model.predict(X_test)
```

```python
regression_model.score(X_test,Y_test)
```

```python
from sklearn import metrics


prediction = regression_model.predict(X_test)

metrics.mean_absolute_error(Y_test,prediction)
```

```python
MAE=metrics.mean_absolute_error(Y_test,prediction)
RMSE=np.sqrt(MAE)
RMSE
```
